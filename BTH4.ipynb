4. Tạo một vector với các giá trị như sau: 20, 17, 32, 19, 65
- Tạo một series từ vector ở trên.
- Thêm tên cho các dòng của series vừa tạo.
- Lấy giá trị ở dòng thứ 2.
- Lấy giá trị 3 dòng cuối của serie trên

```python
vector = [20, 17, 32, 19, 65]
series = pd.Series(vector)
series.index = ['A', 'B', 'C', 'D', 'E']
value_second_row = series['B']
last_three_values = series.tail(3)
print("Series:\n", series)
print("\nGiá trị ở dòng thứ 2: ", value_second_row)
print("\n3 dòng cuối của series:\n", last_three_values)
```
